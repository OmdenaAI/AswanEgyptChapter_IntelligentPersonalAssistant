```python
import os
import pandas as pd

# Get the current directory
current_directory = os.getcwd()

# List all files in the current directory
all_files = os.listdir(current_directory)

# Filter files to keep only Excel files
excel_files = [file for file in all_files if file.endswith(('.xls', '.xlsx'))]

# Initialize an empty list to store DataFrames
dfs = []

# Define the desired column order
desired_columns_order = ['contributor name', 'job', 'Input text', 'features summary']

# Loop through each Excel file, read into a DataFrame, and rename columns
for excel_file in excel_files:
    file_path = os.path.join(current_directory, excel_file)
    df = pd.read_excel(file_path)

    # Rename columns based on the desired order
    df = df.rename(columns=dict(zip(df.columns, desired_columns_order)))

    # Append the modified DataFrame to the list
    dfs.append(df)

```

```python
# Concatenate all DataFrames into a single DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)

# Display the concatenated DataFrame
```

```python
#convert feature summary column string back to json format and extract each feature to it's column
i = 0
for  schedule in schedules_2:
    
    
    features_dict = json.loads(schedule)
    
    if ' time' in features_dict:
        features_dict["time"] = features_dict.pop(" time")

    
    features_dict['time'] = features_dict.get('time', None)
    
    concatenated_df = pd.concat([concatenated_df, pd.DataFrame(columns=features_dict.keys())])


    
    for column, value in features_dict.items():
        
        concatenated_df.at[i, column] = value
            
    i += 1
```
